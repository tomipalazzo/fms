```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Ingreso el DataSet

```python
fms_arg_2018 = pd.read_csv('/home/tomi/Tomi/repos/fms/dataSets/fms_argentina_2018.csv', index_col='Fecha')
```

```python
fms_arg_2018
```

```python
def puntosPorJornada(df, competidor):
    puntos = [df[competidor].loc[1]] #Caso Base
    for i in range(2, df.shape[0] + 1):
        puntos.append(df[competidor].loc[i] - df[competidor].loc[i-1])
    
    return puntos
        
```

```python
puntosPorJornada(fms_arg_2018, 'trueno')
```

```python
fechas = fms_arg_2018.index
puntos_trueno = puntosPorJornada(fms_arg_2018, 'trueno')
puntos_mks = puntosPorJornada(fms_arg_2018, 'mks')



#Grafico
plt.figure(figsize=(16,6))
#Trueno
plt.scatter(fechas, puntos_trueno, s=80, color='cyan')
plt.plot(fechas, puntos_trueno, label='Trueno', color='cyan')

#MKS
plt.scatter(fechas, puntos_mks, s=80)
plt.plot(fechas, puntos_mks, label='MKS')



plt.title('Trueno vs MKS', fontsize=20)
plt.xlabel('Jornada', fontsize=20)
plt.ylabel('Puntos x fecha', fontsize=20)


plt.grid(fontsize=16)

plt.show()
```
