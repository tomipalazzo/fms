```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fms_arg_2018 = pd.read_csv('/home/tomi/Tomi/repos/fms/dataSets/fms_argentina/fms_argentina_2018.csv', index_col='Fecha')
fms_arg_2019 = pd.read_csv('/home/tomi/Tomi/repos/fms/dataSets/fms_argentina/fms_argentina_2019.csv', index_col='Fecha')
fms_arg_2020 = pd.read_csv('/home/tomi/Tomi/repos/fms/dataSets/fms_argentina/fms_argentina_2020.csv', index_col='Fecha')
```

```python
fms_arg_2020
```

```python
def puntosPorJornada(df, competidor):
    puntos = [df[competidor].loc[1]] #Caso Base
    for i in range(2, df.shape[0] + 1):
        puntos.append(df[competidor].loc[i] - df[competidor].loc[i-1])
    
    return puntos
        
```

## Puntos Trueno vs MKS 2018

Se puede observar que trueno en la mayoria de las jornadas fue superior a MKS, lo que se destaca es que Trueno descendio y MKS obtuvo un lugar en el podio

```python
fechas = fms_arg_2018.index
puntos_trueno = puntosPorJornada(fms_arg_2018, 'trueno')
puntos_mks = puntosPorJornada(fms_arg_2018, 'mks')



#Grafico
plt.figure(figsize=(16,8))
#Trueno
plt.scatter(fechas, puntos_trueno, s=80)
plt.plot(fechas, puntos_trueno, label='Trueno')

#MKS
plt.scatter(fechas, puntos_mks, s=80)
plt.plot(fechas, puntos_mks, label='MKS')



plt.title('Trueno vs MKS - 2018', fontsize=20)
plt.xlabel('Jornada', fontsize=20)
plt.ylabel('Puntos por jornada', fontsize=20)

plt.legend(fontsize=16)
plt.grid()

plt.savefig('truenovsMKS-2018.png')


plt.show()
```

## Comparo a Stuart 2018 vs Stuart 2019
Se puede ver como Stuart pudo aumentar su nivel en la competencia

```python
stuart_18 = puntosPorJornada(fms_arg_2018, 'stuart')
stuart_19 = puntosPorJornada(fms_arg_2019, 'stuart')

#Grafico
plt.figure(figsize=(15,7))
#2018
plt.scatter(fechas, stuart_18, s=80)
plt.plot(fechas, stuart_18, label='2018')

#2019
plt.scatter(fechas, stuart_19, s=80)
plt.plot(fechas, stuart_19, label='2019')



plt.title('Stuart 2018 vs. Stuart 2019', fontsize=20)
plt.xlabel('Jornada', fontsize=20)
plt.ylabel('Puntos por Jornada', fontsize=20)

plt.legend(fontsize=16)
plt.grid()
plt.savefig('stuart_18vs19.png')

plt.show()
```
